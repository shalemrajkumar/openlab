Golden search

```python
# initializing a and b as bounds
def golden_section_search(a, b, f, epsilon=1e-5):
    gr = (np.sqrt(5) - 1) / 2
    
    
    x1 = a + (gr * (b-a))
    x2 = b - (gr * (b-a))
    
    for i in range(1000):

        if b-a < epsilon:
            break
    
        if f(x1) < f(x2):
    
            a = x2
            x2 = x1
            x1 = a + gr*(b-a)
    
        else:
    
            b = x1
            x1 = x2
            x2 = b - gr*(b-a)

    return a, b
```

```python
a, b = -5, 5
```

```python
golden_section_search(a, b, f)
```
